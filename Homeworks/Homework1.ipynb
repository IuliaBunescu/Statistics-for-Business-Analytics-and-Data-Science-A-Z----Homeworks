#### Men Height Normal Distribution
1000 observations, with a mean of $\mu = 175.5cm$  and standard deviation of $ \sigma = 7.4cm $

```python
from my_functions import normal_distribution
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sb

miu = 175.5
sigma = 7.4
N = 1000

data = np.linspace(-4*sigma+miu,4*sigma+miu, N)
sorted_data = np.sort(data)

density = normal_distribution(miu, sigma, sorted_data)


sb.set_style('whitegrid')
sb.lineplot(x = sorted_data, y = density , color = 'black')
plt.xlabel('Men Heights')
plt.ylabel('Probability Density')
plt.title('Men Heights in US')
```

Identify the minimum height of 2.2% o the tallest people in that population.

```python
# We use the z-score: z = (x-miu) / sigma, 
P_of_z = 0.978
z = 2 #approximatly, form the z-score table
x = z * sigma + miu

print('The minimum height for a male to be in the top 2.2% is {}'.format(x))
```
